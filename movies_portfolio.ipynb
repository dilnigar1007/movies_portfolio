```python
    #  Import and read the title-ratings.csv
    import pandas as pd 
    ratings_df = pd.read_csv("title-ratings.csv")
    ratings_df.head()
```

```python
ratings_df.sort_values(by=['averageRating'], ascending=False)
```

```python
sorted_df = ratings_df.sort_values(by=['averageRating'], ascending=False)
sorted_df.head(10)
```

```python
#  Import and read the title-metadata.csv
title_df = pd.read_csv("title-metadata.csv")
title_df.head()
```

```python
five_movies_combined_df = pd.merge(sorted_df, title_df, how="inner", on=["tconst"])
five_movies_combined_df.head()
```

```python
top_five_movies_df = five_movies_combined_df.head(5)
top_five_movies_df
```

```python
top_five_movies_df.to_csv("top_five_movies.csv")
```
