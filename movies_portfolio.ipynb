```python
    #  Import and read the title-metadata.csv
    import pandas as pd 
    data_df = pd.read_csv("title-metadata.csv")
    data_df.head()
```

```python
data_2_df = data_df[data_df['startYear'].notna()]
new_df = data_2_df[data_2_df.startYear!= "\\N"]
new_df.head()
```

```python
new_df['startYear'] = new_df['startYear'].astype('int')
```

```python
movies_df = new_df[new_df['startYear'] > 2010]
movies_df.head()
```

```python
movies_df.to_csv("new movies.csv")
```

```python
#  Import and read the title-ratings.csv
ratings_df = pd.read_csv("title-ratings.csv")
ratings_df.head()
```

```python
sorted_df = ratings_df.sort_values(by=['numVotes', 'averageRating'], ascending=False)
sorted_df.head(10)
```

```python
top_videos_df= pd.merge(sorted_df, movies_df, how="inner", on=["tconst"])
top_videos_df.head()
```

```python
top_movies_df = top_videos_df[top_videos_df['titleType'] == "movie"]
top_movies_df.head()
```

```python
top_five_movies_df = top_movies_df.head(5)
top_five_movies_df
```

```python
top_five_movies_df.to_csv("top_five_movies.csv")
```
